```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt


customers = pd.read_csv(path+"/olist_customers_dataset.csv")
geolocation = pd.read_csv(path+"/olist_geolocation_dataset.csv")
order_items = pd.read_csv(path+"/olist_order_items_dataset.csv")
payments = pd.read_csv(path+"/olist_order_payments_dataset.csv")
reviews = pd.read_csv(path+"/olist_order_reviews_dataset.csv")
orders = pd.read_csv(path+"/olist_orders_dataset.csv")
products = pd.read_csv(path+"/olist_products_dataset.csv")
product_category = pd.read_csv(path+"/product_category_name_translation.csv")
```

1.Crescimento e Receita • Evolução mensal de pedidos, receita e ticket médio

```python
order_payments = orders.merge(payments, on='order_id')
order_customer = order_payments.merge(customers, on='customer_id')
order_all = order_customer.merge(order_items, on='order_id')
order_all.head()
```

```python
import pandas as pd
from pyspark.sql import SparkSession

# Initialize SparkSession
spark = SparkSession.builder.appName("OlistDataAnalysis").getOrCreate()

order_all['order_approved_at'] = pd.to_datetime(order_all['order_approved_at'])
order_all['anomes_approved']= order_all['order_approved_at'].dt.to_period('M').astype(str) # Convert to string here
order_receita = order_all.groupby(['anomes_approved','customer_state']).agg({'payment_value': 'sum'}).reset_index()

df_order_receita=spark.createDataFrame(order_receita)
df_order_receita.createOrReplaceTempView("tmp_order_receita")
display(df_order_receita)
```

```python
spark.sql("select * from tmp_order_receita where anomes_approved = '2017-10'").show()
```

```python
order_evolu = order_all.groupby(['anomes_approved']).agg(total_revenue=('payment_value','sum'), order_count=('order_id','count'))
order_evolu['ticket_medio'] = order_evolu['total_revenue']/order_evolu['order_count']

order_evolu.head()
```

```python
# Calculate top 10 products by total revenue
top_products_revenue = order_products.groupby('product_id').agg(total_revenue=('price', 'sum')).nlargest(10, 'total_revenue').reset_index()
print("Top 10 Products by Total Revenue:")
print(top_products_revenue)

# Calculate top 10 sellers by total revenue
top_sellers_revenue = order_products.groupby('seller_id').agg(total_revenue=('price', 'sum')).nlargest(10, 'total_revenue').reset_index()
print("\nTop 10 Sellers by Total Revenue:")
print(top_sellers_revenue)

# Calculate top 10 products by number of sales
top_products_sales = order_products.groupby('product_id').agg(total_sales=('order_id', 'count')).nlargest(10, 'total_sales').reset_index()
print("\nTop 10 Products by Number of Sales:")
print(top_products_sales)

# Calculate top 10 sellers by number of sales
top_sellers_sales = order_products.groupby('seller_id').agg(total_sales=('order_id', 'count')).nlargest(10, 'total_sales').reset_index()
print("\nTop 10 Sellers by Number of Sales:")
print(top_sellers_sales)
```

```python
product_customer_category = product_customer.merge(product_category, on='product_category_name')

category_region_revenue = product_customer_category.groupby(['product_category_name_english', 'customer_state']).agg(total_revenue=('price', 'sum')).reset_index()

print("Top 10 product categories by total revenue across all states:")
print(category_region_revenue.groupby('product_category_name_english')['total_revenue'].sum().nlargest(10))

print("\nTop 5 states for each of the top 3 product categories:")
top_3_categories = category_region_revenue.groupby('product_category_name_english')['total_revenue'].sum().nlargest(3).index

for category in top_3_categories:
    print(f"\nCategory: {category}")
    print(category_region_revenue[category_region_revenue['product_category_name_english'] == category].nlargest(5, 'total_revenue'))
```
